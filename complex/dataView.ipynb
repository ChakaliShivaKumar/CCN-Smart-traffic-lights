```python
import xml.etree.ElementTree as ET

# Parse the XML file
tree = ET.parse("complex_tripinfo.xml")
root = tree.getroot()

# Loop through each tripinfo element and print the required information
for trip in root.findall("tripinfo"):
    # Get various attributes from each tripinfo element
    vehicle_id = trip.get("id")
    duration = trip.get("duration")
    depart = trip.get("depart")
    arrival = trip.get("arrival")
    waiting_time = trip.get("waitingTime")
    route_length = trip.get("routeLength")
    
    # Print vehicle details
    print(f"Vehicle {vehicle_id} - Duration: {duration}s")
    print(f"  Depart: {depart}s, Arrival: {arrival}s")
    print(f"  Waiting Time: {waiting_time}s, Route Length: {route_length}m")
    print("----")
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

# Parse the XML file
tree = ET.parse("complex_tripinfo.xml")  # Replace with your file path
root = tree.getroot()

# Extract all attributes from each <tripinfo> element
data = []
for trip in root.findall("tripinfo"):
    trip_data = trip.attrib  # Get all attributes as a dictionary
    data.append(trip_data)

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df.head())  # Shows the first few rows
df.to_csv("tripinfo_data.csv", index=False)  # Save to CSV (optional)
```
